# ðŸŒŸ Exercise 2: Artificial Neural Networks (ANNs)

Instructions:

1. Draw (or use a diagram tool) to create a simple artificial neural network (ANN) with:

An input layer (3 neurons)
One hidden layer (4 neurons)
An output layer (2 neurons)

2. Label the components of the network: neurons, weights, biases, activation function, and layers.

3. Write a short description (3-5 sentences) explaining how information flows through the network.

```python
import matplotlib.pyplot as plt
import networkx as nx
import matplotlib.patches as mpatches

# Define the number of neurons per layer
layers = {
    "input": 3,
    "hidden": 4,
    "output": 2
}

# Define x-positions for layers
layer_x = {"input": 0, "hidden": 1, "output": 2}

# Initialize the graph
G = nx.DiGraph()
positions = {}
labels = {}
colors = []

# Create nodes with positions and labels
for layer_name, num_nodes in layers.items():
    for i in range(num_nodes):
        node_id = f"{layer_name}_{i+1}"
        G.add_node(node_id)
        positions[node_id] = (layer_x[layer_name], -i)
        labels[node_id] = node_id
        if layer_name == "input":
            colors.append("lightgray")
        elif layer_name == "hidden":
            colors.append("skyblue")
        else:
            colors.append("lightgreen")

# Create weighted connections (edges) between layers
for i in range(layers["input"]):
    for j in range(layers["hidden"]):
        G.add_edge(f"input_{i+1}", f"hidden_{j+1}")

for i in range(layers["hidden"]):
    for j in range(layers["output"]):
        G.add_edge(f"hidden_{i+1}", f"output_{j+1}")

# Create legend patches
legend_patches = [
    mpatches.Patch(color='lightgray', label='Input Neurons'),
    mpatches.Patch(color='skyblue', label='Hidden Neurons\n(with Bias & Activation Function)'),
    mpatches.Patch(color='lightgreen', label='Output Neurons'),
    mpatches.Patch(color='black', label='Weights (Connections)')
]

# Plot the network
plt.figure(figsize=(12, 7))
nx.draw(G, pos=positions, with_labels=True, labels=labels,
        node_color=colors, node_size=1500, font_size=10,
        edge_color='black', arrows=True)

# Annotate layers
plt.text(-0.2, 0.1, "Input Layer", fontsize=12, fontweight='bold')
plt.text(0.8, 0.1, "Hidden Layer\n(+ Bias + Activation)", fontsize=12, fontweight='bold')
plt.text(2.1, 0.1, "Output Layer", fontsize=12, fontweight='bold')

# Add legend
plt.legend(handles=legend_patches, loc='upper center', bbox_to_anchor=(0.5, -0.08), ncol=2, fontsize=10)

# Final touches
plt.title("Annotated Artificial Neural Network (3-4-2)\nLabeled Components", fontsize=14)
plt.axis("off")
plt.subplots_adjust(top=0.85, bottom=0.2)  # Avoid tight_layout warning
plt.show()

```

In this artificial neural network (ANN), data flows from the input layer to the hidden layer and finally to the output layer. Each neuron in a layer is connected to every neuron in the next layer through weighted edges. These weights, combined with biases, determine the strength and direction of the signal being passed. At each neuron, an activation function (such as ReLU or sigmoid) is applied to the weighted sum of inputs to introduce non-linearity. This allows the network to learn complex patterns and make predictions based on the processed input data.

## ðŸŒŸ Exercise 3: Creating the Dataset and Visualizing the Data

Objective: Generate a dataset with noise and visualize it.

Instructions:

1. Import the necessary libraries: numpy, matplotlib.pyplot, and mean_squared_error from sklearn.metrics.

2. Create a set of 20 points using the formula ( y = -x^2 ), adding normally distributed noise with a mean of 0 and a standard deviation of 0.05 using np.random.normal.

Hint:

3. Plot the generated points using a scatter plot to visualize the dataset with noise.

4. Split the dataset into a training set (first 12 points) and a test set (last 8 points).



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Set the seed for reproducibility
np.random.seed(0)

# Generate x values from -1 to 1 with step 0.1
x = np.arange(-1, 1, 0.1)

# Generate noisy y values based on the equation y = -x^2 + noise
y = -x**2 + np.random.normal(0, 0.05, len(x))

# Visualize the dataset with noise using a scatter plot
plt.scatter(x, y, color='blue', label='Noisy Data')
plt.title('Noisy Dataset: y = -x^2 + noise')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

# Split the dataset into training (first 12 points) and test (last 8 points)
x_train = x[:12]
y_train = y[:12]
x_test = x[12:]
y_test = y[12:]

# Output to confirm the split
print("Training Set:")
print("x_train:", x_train)
print("y_train:", y_train)
print("\nTest Set:")
print("x_test:", x_test)
print("y_test:", y_test)
```

## ðŸŒŸ Exercise 4: Fitting Polynomial Models of Different Degrees

Instructions:

1. Define a function polynomial_fit that takes a degree as input and returns a polynomial that best fits the training data.

2. Define another function plot_polyfit that plots the training set, test set, and the fitted polynomial curve for a given degree.

3. Use plot_polyfit to visualize the fit for polynomials of degree 1, 7, and 11.

4. Observe how the model behaves on the training and test sets as the degree of the polynomial increases.

```python
def polynomial_fit(degree):
    return np.poly1d(np.polyfit(x_train, y_train, degree))

def plot_polyfit(degree):
    poly = polynomial_fit(degree)

    # Fresh figure for each plot
    plt.figure(figsize=(8, 5))

    # Scatter plots for clarity
    plt.scatter(x_train, y_train, color='blue', label='Training Data')
    plt.scatter(x_test, y_test, color='red', label='Test Data')

    # Polynomial curve
    x_range = np.linspace(-1, 1, 100)
    plt.plot(x_range, poly(x_range), color='green', label=f'Degree {degree}')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Polynomial Fit (Degree {degree})')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_polyfit(1)
```

```python
plot_polyfit(7)
```

```python
plot_polyfit(11)
```

```python
plot_polyfit(2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error


# Store results
results = []

for degree in range(1, 12):
    poly = polynomial_fit(degree)

    # Predict on training and test sets
    y_train_pred = poly(x_train)
    y_test_pred = poly(x_test)

    # Calculate RMSE
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

    # Store results
    results.append((degree, rmse_train, rmse_test))

# Convert results to arrays for easy plotting
degrees = [r[0] for r in results]
train_rmse = [r[1] for r in results]
test_rmse = [r[2] for r in results]

# Plotting RMSE
plt.figure(figsize=(10, 6))
plt.plot(degrees, train_rmse, marker='o', label='Training RMSE')
plt.plot(degrees, test_rmse, marker='o', label='Test RMSE')
plt.yscale('log')
plt.xlabel('Polynomial Degree')
plt.ylabel('RMSE (log scale)')
plt.title('Training vs Test RMSE for Polynomial Degrees 1 to 11')
plt.legend()
plt.grid(True)
plt.show()

# Identify best degree based on test RMSE
best_degree = degrees[np.argmin(test_rmse)]
print(f"âœ… Best degree based on Test RMSE: {best_degree}")
```
