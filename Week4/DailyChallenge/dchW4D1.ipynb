```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Data Preparation:

Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities. The temperatures should range from -5 to 35 degrees.

Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns.

Hint 1: Use np.random.uniform(low, high, size) to generate the temperature data.

```python
data = np.random.uniform(low=-5, high=35, size=120)
data = np.round(data, 2)
data = data.reshape(10, 12)
data
```

Hint 2: Create a DataFrame using pd.DataFrame(data, index, columns) with appropriate index and columns.

```python
# creat dataframe for 10 cties and 12 months
cities = ["city1", "city2", "city3", "city4", "city5", "city6", "city7", "city8", "city9", "city10"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
df = pd.DataFrame(data, index=cities, columns=months)
df
```

# 2. Data Analysis:

Calculate the annual average temperature for each city using DataFrame.mean(axis).

```python
for city in df.index:
    print(f"{city}: {df.loc[city].mean()}")
```

Identify the city with the highest and lowest average temperature for the year using idxmax() and idxmin() methods.

```python
max_city = df.mean(axis=1).idxmax()
min_city = df.mean(axis=1).idxmin()
print(f"The city with the highest average temperature is {max_city} with {df.mean(axis=1).max()} degrees.")
print(f"The city with the lowest average temperature is {min_city} with {df.mean(axis=1).min()} degrees.")
```

# 3. Data Visualization:

```python
#create monthly temperature trends for each city
plt.plot(df)
plt.xlabel("Month")
plt.ylabel("Temperature (Celsius)")
plt.title("Monthly Temperature Trends for Each City")
plt.legend(df.index)
```
