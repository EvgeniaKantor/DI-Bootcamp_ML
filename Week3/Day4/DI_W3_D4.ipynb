```python
# Visualize the sales of five products (e.g., Product A, Product B, etc.) in a bar chart.
# Customize the chart with different colors and patterns.
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [100, 150, 120, 200, 180]
colors = ['red', 'green', 'blue', 'orange', 'purple']
plt.bar(products, sales, color=colors)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Sales of Products')
plt.show()
```

```python
# Generate a dataset of 500 random numbers following a normal distribution.
# Plot the histogram and experiment with different bin sizes and colors.
data = np.random.randn(500)
plt.hist(data, bins=100, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers')
plt.show()
```

```python
# Create a pie chart showing the percentage distribution of time spent on different daily activities (e.g., work, sleep, leisure, exercise, etc.).
# Use custom colors and add a shadow effect.
activities = ['Work', 'Sleep', 'Leisure', 'Exercise', 'Other']
time_spent = [30, 20, 15, 10, 15]
explode = (0.1, 0, 0, 0, 0)
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff6666']
plt.pie(time_spent, labels=activities, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True)
plt.title('Daily Activity Distribution')
plt.show()
```

# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample data
data = sns.load_dataset('tips')
sns.histplot(data['total_bill'], bins=20, color='skyblue')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
sns.kdeplot(data['total_bill'], shade=True, color='purple')
plt.title('KDE of Total Bill')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='scatter', color='green')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
sns.barplot(x='class', y='survived', data=titanic, palette='Blues')
plt.title('Survival Rate by Class')
plt.show()
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='Set3')
plt.title('Age Distribution by Class')
plt.show()
```

```python
sns.violinplot(x='class', y='age', data=titanic, palette='muted')
plt.title('Age Distribution by Class')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=data, aspect=1.5, height=5)
plt.title('Scatter Plot with Regression Line')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='reg', color='orange')
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species', palette='Dark2')
plt.show()
```

```python
iris.head()
```

```python
numerical_columns = iris.select_dtypes(include=['number'])
correlation_matrix = numerical_columns.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
data = sns.load_dataset('tips')
g = sns.FacetGrid(data, col='time', row='smoker', margin_titles=True)
g.map(sns.scatterplot, 'total_bill', 'tip', color='red')
plt.show()
```

```python
iris = sns.load_dataset('iris')
g = sns.PairGrid(iris, hue='species')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap='Blues')
g.map_diag(sns.histplot, kde=True, color='teal')
g.add_legend()
plt.show()
```

```python
sns.set_context('talk')
sns.set_style('whitegrid')
sns.histplot(data['total_bill'], color='darkblue', bins=15)
plt.title('Advanced Aesthetic Customization')
plt.show()
```

```python
# 1. Create a histogram of customer spending from the tips dataset. Customize the bins and color.
data = sns.load_dataset('tips')
sns.histplot(data['total_bill'], bins=20, color='skyblue')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
# 2. Use pairplot on the iris dataset with different markers for species.
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species', markers=['o', 's', 'D'])
plt.show()
```

```python
# 3. Visualize correlations in the iris dataset using a heatmap.
iris = sns.load_dataset('iris')
numerical_columns = iris.select_dtypes(include=['number'])
correlation_matrix = numerical_columns.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

```python
# 4. Build a FacetGrid using the titanic dataset to compare survival rates by age across class and gender.
titanic = sns.load_dataset('titanic')
# Drop rows where 'age' is missing
titanic = titanic.dropna(subset=['age'])

# Create FacetGrid
g = sns.FacetGrid(titanic, col='class', row='sex', margin_titles=True)

# Map scatterplot with hue for survival
g.map_dataframe(sns.scatterplot, x='age', y='survived', hue='survived', palette='viridis', alpha=0.7)

# Show plot
plt.show()
```
