```python
import pandas as pd
data = pd.read_csv('2019.csv')
data.head()
```

# Address missing values and adjust data types as necessary

**Overall rank**: This column represents the global ranking of each country or region based on the happiness or well-being index.

**Country or region**: The name of the country or region being assessed.

**Score**: The happiness score of the country or region, which is based on various factors contributing to the well-being of its citizens. This score is usually out of 10 and is used to rank countries in terms of overall life satisfaction.

**GDP per capita**: This column shows the Gross Domestic Product (GDP) per person in the country or region. A higher GDP per capita typically indicates a wealthier nation with better resources for its citizens.

**Social support**: This measures the perceived level of social support available to people in the country. Social support is generally defined as having someone to count on in times of trouble, such as family or friends.

**Healthy life expectancy**: This column shows the average number of years a person can expect to live in good health. It takes into account factors like healthcare quality, access to medical care, and lifestyle choices.

**Freedom to make life choices:** This measures how free people feel to make their own decisions in life, such as career choices, marriage, and other personal decisions. It's an indicator of the level of autonomy people have in their lives.

**Generosity:** This is an index measuring the level of generosity within a country, usually derived from measures of donations and volunteering behaviors in the population.

**Perceptions of corruption:** This column captures the level of corruption perceived by the citizens in the country or region.

# Employ a Matplotlib scatter plot to investigate the relationship between ‘Social support’ and ‘Score’.

```python
import matplotlib.pyplot as plt

plt.scatter(data['Social support'], data['Score'])
plt.xlabel('Social support')
plt.ylabel('Score')
plt.title('Relationship between Social support and Score')
plt.show()
```

# Create a Matplotlib subplot that compares ‘GDP per Capita’ and ‘Healthy Life Expectancy’ across different regions.

```python
# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(10, 21))

# Bar plot for 'GDP per Capita'
ax1.barh(data['Country or region'], data['GDP per capita'], color='skyblue', label='GDP per Capita')
ax1.set_xlabel('GDP per Capita', color='skyblue')
ax1.set_ylabel('Country/Region', color='black')
ax1.tick_params(axis='x', labelcolor='skyblue')

# Create a second y-axis for the line plot
ax2 = ax1.twiny()

# Line plot for 'Healthy Life Expectancy'
ax2.plot(data['Healthy life expectancy'], data['Country or region'], color='green', marker='o', label='Healthy Life Expectancy')
ax2.set_xlabel('Healthy Life Expectancy', color='green')
ax2.tick_params(axis='x', labelcolor='green')

# Set title for the plot
plt.title('Comparison of GDP per Capita and Healthy Life Expectancy')

# Show the plot
plt.tight_layout()
plt.show()
```
